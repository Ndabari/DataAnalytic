Building visualizations

```python
pyplot.bar(housing_df['bedrooms'], housing_df['price'])
pyplot.xlabel('Bedrooms')
pyplot.ylabel('Price in Millions')
pyplot.title('Bedrooms Vs Price in Millions')
pyplot.show()
```

Horizontal bar graph (barh)

```python
pyplot.barh(housing_df['bedrooms'], housing_df['price'])
pyplot.xlabel('Bedrooms')
pyplot.ylabel('Price in Millions')
pyplot.title('Bedrooms Vs Price in Millions')
pyplot.show()
```

```python
pyplot.scatter(housing_df['bathrooms'], housing_df['price'])
pyplot.xlabel('Bedrooms')
pyplot.ylabel('Price in Millions')
pyplot.title('Bedrooms Vs Price in Millions')
pyplot.show()
```
